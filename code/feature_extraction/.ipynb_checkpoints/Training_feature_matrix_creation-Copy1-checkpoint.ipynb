```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
#for now in <cell_line_specific_input_train_dir_MACCS_target> files will contain MACCSKeys and target features.
cell_line_specific_input_train_dir_MACCS_target = "/data/tasnina/drug-synergy-prediction/inputs/train/MACCS_Target/"

#for now in <cell_line_specific_input_train_dir_only_MACCS> files will contain only MACCSKeys feature.
cell_line_specific_input_train_dir_only_MACCS = "/data/tasnina/drug-synergy-prediction/inputs/train/only_MACCS/"

input_train_dir = '/data/tasnina/drug-synergy-prediction/inputs/train/'
single_drug_training_maccskeys_target_feature_file_path = input_train_dir + 'single_drugs_maccskeys_target_features.tsv'
single_drug_training_maccskeys_feature_file_path = input_train_dir + 'single_drugs_maccskeys_features.tsv'

# drug_training_feature_file_path = input_train_dir + 'single_drugs_maccskeys_target_features.tsv'
training_label_synergy_score_file_path =  input_train_dir + 'synergy_labels.tsv'
```

```python
synergy_df = pd.read_csv(training_label_synergy_score_file_path, sep='\t', \
                         dtype = {'Drug1_pubchem_cid': str ,'Drug2_pubchem_cid': str,\
                         'Cell_line':str, 'Loewe': np.float64, 'Bliss': np.float64, 'ZIP': np.float64},\
                         index_col = None)

cell_lines = synergy_df['Cell_line'].unique()
```

```python
drug_maccskeys_target_feature_df = pd.read_csv(single_drug_training_maccskeys_target_feature_file_path, sep='\t', index_col=None)
drug_maccskeys_target_feature_df['pubchem_cid'] = drug_maccskeys_target_feature_df['pubchem_cid'].astype(str)

for cell_line in cell_lines:
    feature_file = cell_line_specific_input_train_dir_MACCS_target + cell_line + '_features.tsv'
    label_file =  cell_line_specific_input_train_dir_MACCS_target + cell_line + '_labels.tsv'
    os.makedirs(os.path.dirname(feature_file), exist_ok=True)
    
#     if(not os.path.exists(feature_file) or not os.path.exists(label_file)):
    try:
        cell_line_sdf = synergy_df[synergy_df['Cell_line']==cell_line]

        cell_line_specific_labels = cell_line_sdf['Loewe'].reset_index().drop('index',axis=1)
        cell_line_specific_labels.to_csv(label_file, sep='\t', index=False)
    #     print(cell_line_specific_labels)


        #fisrt take only the rows from drug_maccskeys_target_feature_df which are in 'Drug1_pubchem_cid' in synergy_df and sort
        #according to 'Drug1_pubchem_cid'
        #then  take only the rows from drug_maccskeys_target_feature_df which are in 'Drug2_pubchem_cid' in synergy_df and sort
        #then combine these two dataframe
        drug1_feature_df = drug_maccskeys_target_feature_df[drug_maccskeys_target_feature_df['pubchem_cid'].isin(cell_line_sdf['Drug1_pubchem_cid'])]               
        drug2_feature_df = drug_maccskeys_target_feature_df[drug_maccskeys_target_feature_df['pubchem_cid'].isin(cell_line_sdf['Drug2_pubchem_cid'])]
#         print(drug1_feature_df.shape)
#         print(drug2_feature_df.shape)

        drug1_feature_df = drug1_feature_df.set_index('pubchem_cid').reindex(list(cell_line_sdf['Drug1_pubchem_cid']))
        drug1_feature_df.reset_index(inplace=True)

        drug2_feature_df = drug2_feature_df.set_index('pubchem_cid').reindex(list(cell_line_sdf['Drug2_pubchem_cid']))
        drug2_feature_df.reset_index(inplace=True)


        drug1_feature_df = drug1_feature_df.drop('pubchem_cid',axis=1)
        drug2_feature_df = drug2_feature_df.drop('pubchem_cid', axis=1)

        cell_specific_feature_df = drug1_feature_df.join(drug2_feature_df, lsuffix = '_1', rsuffix = '_2')
        cell_specific_feature_df.to_csv(feature_file,sep='\t', index=False)

        print(cell_specific_feature_df.shape)
    #     print(cell_specific_feature_df.head())


        # print(drug1_feature_df.shape)
        # print(drug2_feature_df.shape)



        # print(cell_line_sdf['Drug1_pubchem_cid'])
        # print(drug1_feature_df.head(), drug2_feature_df.head())
    except:
        print('file not created for: ', cell_line)


    
```

```python
drug_maccskeys_feature_df = pd.read_csv(single_drug_training_maccskeys_feature_file_path, sep='\t', index_col=None)
drug_maccskeys_feature_df['pubchem_cid'] = drug_maccskeys_feature_df['pubchem_cid'].astype(str)


for cell_line in cell_lines:
    feature_file = cell_line_specific_input_train_dir_only_MACCS + cell_line + '_features.tsv'
    label_file =  cell_line_specific_input_train_dir_only_MACCS + cell_line + '_labels.tsv'
    os.makedirs(os.path.dirname(feature_file), exist_ok=True)
    
#     if(not os.path.exists(feature_file) or not os.path.exists(label_file)):
    try:
        cell_line_sdf = synergy_df[synergy_df['Cell_line']==cell_line]

        cell_line_specific_labels = cell_line_sdf['Loewe'].reset_index().drop('index',axis=1)
        cell_line_specific_labels.to_csv(label_file, sep='\t', index=False)
    #     print(cell_line_specific_labels)


        #fisrt take only the rows from drug_feature_df which are in 'Drug1_pubchem_cid' in synergy_df and sort
        #according to 'Drug1_pubchem_cid'
        #then  take only the rows from drug_feature_df which are in 'Drug2_pubchem_cid' in synergy_df and sort
        #then combine these two dataframe
        drug1_feature_df = drug_maccskeys_feature_df[drug_maccskeys_feature_df['pubchem_cid'].isin(cell_line_sdf['Drug1_pubchem_cid'])]               
        drug2_feature_df = drug_maccskeys_feature_df[drug_maccskeys_feature_df['pubchem_cid'].isin(cell_line_sdf['Drug2_pubchem_cid'])]
#         print(drug1_feature_df.shape)
#         print(drug2_feature_df.shape)

        drug1_feature_df = drug1_feature_df.set_index('pubchem_cid').reindex(list(cell_line_sdf['Drug1_pubchem_cid']))
        drug1_feature_df.reset_index(inplace=True)

        drug2_feature_df = drug2_feature_df.set_index('pubchem_cid').reindex(list(cell_line_sdf['Drug2_pubchem_cid']))
        drug2_feature_df.reset_index(inplace=True)


        drug1_feature_df = drug1_feature_df.drop('pubchem_cid',axis=1)
        drug2_feature_df = drug2_feature_df.drop('pubchem_cid', axis=1)

        cell_specific_feature_df = drug1_feature_df.join(drug2_feature_df, lsuffix = '_1', rsuffix = '_2')
        cell_specific_feature_df.to_csv(feature_file,sep='\t', index=False)

        print(cell_specific_feature_df.shape)

    except:
        print('file not created for: ', cell_line)


    


        
```
